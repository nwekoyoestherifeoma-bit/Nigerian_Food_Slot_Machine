```python
# Nigerian Food Slot Machine
# My first Python game!
import random
```

```python
balance = 50000
currency = "₦"
min_bet = 100
max_bet = 500
foods = ["Jollof", "Suya", "MoiMoi", "Egusi", "Peppersoup"]
```

```python
def spin():
    return [random.choice(foods) for i in range (3)]
```

```python
print("=" * 40)
print(" NIGERIAN FOOD SLOT MACHINE ")
print(" Match 3 foods and win 5x! ")
print("=" * 40)
print(f"Starting balance: {currency}{balance}")
```

```python
while balance > 0:
    print(f"\nBalance: {currency}{balance}")
    bet = int(input(f"Place bet ({currency}{min_bet} - {currency}{max_bet}): "))

    if bet < min_bet or bet > max_bet:
        print(f"Bet must be between {currency}{min_bet} and {currency}{max_bet}!")
        continue

    result = spin()
    print("\n Spinning...")
    print(f" {result[0]} | {result[1]} | {result[2]}")

    if result[0] == result[1] == result[2]:
        winnings = bet * 5
        balance += winnings
        print(f" JACKPOT! Chop money! +{currency}{winnings}")
    else:
        balance -= bet
        print(f" No luck! -{currency}{bet}")

    play_again = input("\nSpin again? (yes/no): ")
    if play_again.lower() == "no":
        break

print(f"\nGame over! Final balance: {currency}{balance}")
print("Thanks for playing Nigerian Food Slot Machine!")

```
